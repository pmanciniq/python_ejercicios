```python
votos = pd.read_csv("primary_results.csv")
votos
```

```python
df = votos #df es la forma convencional de llamar a la variable del "data frame"
df
```

```python
df.head(3) #muestra 3 primeras filas
```

```python
df.tail(5) #ultimas 5 lineas
```

```python
df.describe() #muestra estadisticos descriptivos 
```

```python
df.index
```

```python
#se puede filtrar con varias condiciones
df[(df.county=="Manhattan")&(df.party=="Democrat")]
```

```python
#otra forma mas resumida es usando el metodo query
county="Manhattan"

df.query("county==@county and party=='Democrat'")
```

```python
#Ordenar el dataframe segun una columna y ascendente/descendente
df.sort_values(by="votes", ascending=False)
```

```python
#se puede ordenar y agrupar en funcion de columnas y los datos que queremos obtener, por ejemplo:
#por una columna
df.groupby("state").votes.sum()
```

```python
#por dos columnas
df.groupby(["state","party"]).votes.sum()
```

```python
#podemos pasar la funcion apply a una columna.El _abbreviation es para mostrar primera letra.
df.state_abbreviation.apply(lambda s: s[0])
```

```python
df["letra_estado"] = df.state_abbreviation.apply(lambda s: s[0])
```

```python
df.head()
```

```python
#se pueden mostrar los datos de dos tablas, es como un JOIN en SQL
df_pobreza = pd.read_csv("PovertyEstimates.csv")
df_pobreza
```

```python
#muy parecido al JOIN de SQL
df_merged = df.merge(df_pobreza, left_on="fips", right_on="FIPStxt")
df_merged.head()
```
