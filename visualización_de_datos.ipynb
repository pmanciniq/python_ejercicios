```python
import pandas as pd
#importamos pandas
```

```python
df = pd.read_csv("boston_dataset.csv")
df.head()
```

```python
#cambiamos nombre de columnas para hacer mas legible
df = df.rename(columns={
    "TOWN": "CIUDAD",
    "CRIM": "INDICE_CRIMEN",
    "ZN": "PCT_ZONA_RESIDENCIAL",
    "INDUS": "PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO_CHARLES",
    "NOX": "OXIDO_NITROSO_PPM",
    "RM": "N_HABITACIONES_MEDIO",
    "AGE": "PCT_CASAS_40S",
    "DIS_EMPLEO": "DISTANCIA_CENTRO_EMPLEO",
    "RAD": "DIS_AUTOPISTAS",
    "TAX": "CARGA_FISCAL",
    "PTRATIO": "RATIO_PROFESORES",
    "B": "PCT_NEGRA",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
df.head()
```
